```python
import io
import token as token_module
import keyword
def tokenize(code):
    
    code = io.StringIO(code)
    res = tokenize_module.tokenize(lambda:code.readline().encode('utf-8'))
    
    return list(res)
```

```python
def get_all_names(data):
    result = Counter()
    keywords_set = keyword.kwlist
    for solutions in data.values():
        for code in solutions.values():
            try:
                for token in tokenize(code):
                    if token.type == token_module.NAME and token.string not in keywords_set:
                        result[token.string] += 1
            except Exception as e:
                print(e)
#                 print(code)
#                 print("\n\n-------------------------")
    return result
```

```python
def get_set_of_common(counter, n):
    result = set()
    
    for name, cnt in counter.most_common(n):
        result.add(name)
    
    return result

def jaccard(a, b):
    return len(a & b)/len(a | b)
```

```python
def get_scaled_list(pairs, scale):
    result = []
    for data in pairs:
        current = {}
        min_freq = data[0][1] / scale
        for item in data:
            if item[1] >= min_freq:
#                 current.append(item)
                current[item[0]] = item[1]
        
        result.append(current)
    
    return result
            
```

```python
def merge_lists(lists):
    result = {}
    keyset = set()
    for ls in lists:
        keyset |= set(ls.keys())
        
    for key in keyset:
        total_cnt = 0
        for ls in lists:
            total_cnt += ls.get(key, 0)
        
        result[key] = total_cnt
    
    result = sorted(result.items(), key=lambda x: -x[1])
    return result
        
```

```python
lists = get_scaled_list([res_gcj.most_common(), res_anytask.most_common(), res_cf.most_common()], 400)
```

```python
result = merge_lists(lists)
```
